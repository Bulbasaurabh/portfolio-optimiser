```python
def validate_tickers(self, tickers):
        tickers_list = tickers.split(',')
        while i in range(len(tickers_list)):
            ticker=tickers_list[i]
            ticker = ticker.strip()
            if not ticker.isalpha():
                return False
            i=i+1
        return True
```
